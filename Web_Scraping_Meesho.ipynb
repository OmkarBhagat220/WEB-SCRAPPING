```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.meesho.com/jeans-men/pl/3nw')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
name=soup.find("p",class_="sc-eDvSVe gQDOBc NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ejhQZU NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ejhQZU")
name
```

```python
name.text
```

```python
names=soup.find_all("p",class_="sc-eDvSVe gQDOBc NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ejhQZU NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ejhQZU")
names
```

```python
namelist=[]
for i in names:
    namelist.append(i.text)
namelist
```

```python
price=soup.find("h5",class_="sc-eDvSVe dwCrSh")
price
```

```python
price.text
```

```python
prices=soup.find_all("h5",class_="sc-eDvSVe dwCrSh")
prices
```

```python
pricelist=[]
for i in prices:
    pricelist.append(i.text)
pricelist
```

```python
rate=soup.find("span",class_="sc-eDvSVe laVOtN")
rate
```

```python
rate.text
```

```python
rating=soup.find_all("span",class_="sc-eDvSVe laVOtN")
rating
```

```python
ratinglist=[]
for i in rating:
    ratinglist.append(i.text)
ratinglist
```

```python
MRP=soup.find("p", class_="sc-eDvSVe drXXNP sc-jSUZER eSuFsQ sc-jSUZER eSuFsQ")
MRP.text
```

```python
MRPs=soup.find_all("p", class_="sc-eDvSVe drXXNP sc-jSUZER eSuFsQ sc-jSUZER eSuFsQ")
MRPs
```

```python
mrplist=[]
for i in MRPs:
    mrplist.append(i.text)
mrplist
```

```python
import pandas as pd
web_series = pd.DataFrame({})
web_series["Jeans Name"]= namelist
web_series["Price"] = pricelist
web_series["Rating"] = ratinglist
web_series["MRP"] = mrplist
web_series
```

```python
web_series.to_csv("Meesho.csv")
```
